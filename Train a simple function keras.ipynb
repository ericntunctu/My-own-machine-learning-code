```python
## train a stupid function from R to R
### train sin wave
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.datasets import mnist
from keras.models import Sequential 
from keras.layers.core import Dense, Activation
from keras.utils import np_utils
import math
```

```python
## build a trainning function 
n_points=2000
x=np.linspace(0, 10, n_points)
noise = np.random.normal(0,1,n_points)
y=np.sin(x)+0.4*x+0.2*noise
plt.plot(x, y,'o', color='green')
plt.title('trainning function')
plt.xlabel('x')
plt.ylabel('data')
```

```python
## build the model
np.random.seed(0)
model = Sequential()

model.add(Dense(20, input_dim=1, activation='sigmoid'))
model.add(Dense(20, activation='sigmoid'))
model.add(Dense(1))
## model.summary()
model.compile(loss='mean_squared_error',optimizer='sgd')
```

```python
## filling data and training:
X1_train=np.array(x, ndmin=2).T
Y1_train=np.array(y, ndmin=2).T



#### start to train
epochs = 2000;
for epoch in np.arange(1, epochs):
      model.fit(X1_train, Y1_train, batch_size=100)
    
 ## animate       
```

```python
## load results and print
prediction = model.predict(X1_train)
plt.plot(x, y,'o', color='red')
plt.plot(x, np.array(prediction, ndmin=1),'o', color='blue')
```
